```python
fx = 4.5
rho = 2.
mu = 3.
eta_over_rho = -2.
Azero = 4.
```

```python
lagrang = lambda v: mu/2*(v - fx)**2 + rho/2*(v + Azero + eta_over_rho)**2
```

```python
v_opt = fx - rho/(mu + rho)*(fx + Azero + eta_over_rho)
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
fig = plt.figure()
axs = fig.add_axes([0,0,1,1])

axs.plot(np.linspace(v_opt -0.2,v_opt + 0.2,21),lagrang(np.linspace(v_opt - 0.2,v_opt + 0.2,21)))
print(v_opt)
```
