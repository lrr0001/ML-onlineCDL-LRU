```python
import numpy as np
rho = 3.
D = 4.
R= 2.
zprev = 2.5
z = 5.
gamma = 9.
mu = 4.

```

```python
lagrng = lambda x: (mu/2.)*(zprev - D*x)**2 + (rho/2)*((np.sqrt(mu)/R)*(z - x) + gamma/rho)**2
```

```python
x_opt = (R/(rho + D*R*D*R))*(D*R*zprev + rho*(z/R + gamma/(rho*np.sqrt(mu))))
from matplotlib import pyplot as plt
%matplotlib inline
fig = plt.figure()
axs = fig.add_axes([0,0,1,1])

axs.plot(np.linspace(x_opt -0.2,x_opt + 0.2,21),lagrng(np.linspace(x_opt - 0.2,x_opt + 0.2,21)))
print(x_opt)
```
